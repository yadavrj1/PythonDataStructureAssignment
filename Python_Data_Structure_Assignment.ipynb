# Data Types and Structures Questions

1. What are data structures, and why are they important?
  - Data structures ar way to orgnize and store data in a computer so that it can be accessed and used efficiently. They define how data is arranged, related, and accessed. They are containers that hold and manage our data.
  - Importance: efficient data orgnation, optimized operations, data integrity and code readability and maintainability.

2. Explain the difference between mutable and immutable data types with examples.
  - Mutable Data Types: These can be changed after they're created. We can modify, add, or remove elements in-place.
  Example:  List, Dictionary, Set
  - These cannot be changed after creation. Any operation that modifies them will create a new object.
  Example: String, Tuple, Integer, Float, Boolean.

3. What are the main differences between lists and tuples in Python?
  - The main difference between Lists and Tuples. Tuples are mutable their elements can be added, removed, or modified after creation. But Tuples are immutable, meaning their elements cannot be changed once the tuple is created.

4. Describe how dictionaries store data.
  - Dictionaries are unordered collections of data and are represented with curly brackets { } . Dictionaries are mutable and indexed. Dictionaries store data in  key:value format.

5. Why might you use a set instead of a list in Python?
  - Sets store only unique elements. If I need a collection where duplicate values are not allowed, a set automatically handles this. Lists allows duplicates.

6. What is a string in Python, and how is it different from a list?
  - Strings in Python are sequences of characters enclosed in quotes ('' or “”). Lists are ordered collections of items enclosed in square brackets []. Strings are immutable,  they cannot be changed once created, while lists are mutable and can be modified as needed.

7. How do tuples ensure data integrity in Python?
  - Once a tuple is created, its elements cannot be changed, added, or removed. Any attempt to modify a tuple after its creation will result in a TypeError. This fundamental characteristic directly contributes to data integrity.

8. What is a hash table, and how does it relate to dictionaries in Python?
  - A hash table is a data structure that implements an associative array, also known as a dictionary or map. It stores key-value pairs by using a hash function to compute an index into an array of buckets or slots, from which the desired value can be retrieved. This allows for efficient insertion, deletion, and lookup operations.


9. Can lists contain different data types in Python?
  - Yes, lists in Python can contain elements of different data types. This is a key feature of Python lists, making them highly versatile.


10. Explain why strings are immutable in Python.
  - Strings are immutable in Python, meaning their content cannot be changed after they are created.


11. What advantages do dictionaries offer over lists for certain tasks?
  - Dictionaries are fast lookups, key-value associations, and complex data mappings, while lists are ideal for ordered collections, sequential processing, and simple data storage.


12. Describe a scenario where using a tuple would be preferable over a list.
  - I use a list for mutable collection of items where I need to add, remove, or change elements. Lists are more flexible and have more built-in methods, making them ideal for dynamic collections. I use a tuple if I need an immutable collection where the elements won't change after creation.

13. How do sets handle duplicate values in Python?
  - When an attempt is made to add an element that already exists within a set, the set simply ignores the addition and remains unchanged.


14. How does the “in” keyword work differently for lists and dictionaries?
  - In list "in" checks for element(value) and In dictionaries "in" checks for key.

15. Can you modify the elements of a tuple? Explain why or why not?
  - We cannot directly modify the elements of a tuple in Python. Tuples are immutable, meaning their contents cannot be changed after creation. If we attempt to modify a tuple, we will get a TypeError.


16. What is a nested dictionary, and give an example of its use case.

17. Describe the time complexity of accessing elements in a dictionary.
  - Accessing elements in a dictionary typically has a time complexity of O(1) on average. This means that the time it takes to retrieve a value by its key is, on average, constant and doesn't depend on the size of the dictionary. However, in the worst-case scenario, the time complexity can be O(n), where n is the number of elements in the dictionary, due to hash collisions.


18. In what situations are lists preferred over dictionaries?
  - if I need an ordered collection, like a sequence of user actions. On the other hand, when we need quick look-ups, like mapping usernames to user data, dictionaries are preferred.


19. Why are dictionaries considered unordered, and how does that affect data retrieval?
  - A dictionary is considered an unordered collection because it doesn't store items in a specific sequence that corresponds to insertion order. Unlike lists or tuples, we can't rely on the order in which we added elements to a dictionary when retrieving them. Dictionaries are optimized for key-based lookups, not for maintaining order.


20. Explain the difference between a list and a dictionary in terms of data retrieval.
  - A list is an ordered collection of items, whereas a dictionary is an unordered data collection in a key: value pair. Elements from the list can be accessed using the index, while the elements of the dictionary can be accessed using keys.
